```python
# ✅ Function to get predefined or generate if missing
def get_caption(meme_name):
    if meme_name in predefined_captions:
        return predefined_captions[meme_name]
    else:
        # Generate caption using GPT **only for unknown memes**
        prompt = f"Describe the meme '{meme_name}' and when people use it."
        response = caption_generator(prompt, max_new_tokens=20, num_return_sequences=1, do_sample=True, temperature=0.7)
        return response[0]['generated_text'].strip()

# ✅ Apply function to get captions
df["Expanded Captions"] = df["Meme Name"].apply(get_caption)

# ✅ Load BERT Model for Embeddings
bert_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python

# ✅ Convert Enhanced Captions to BERT Embeddings
df["Vector"] = df["Expanded Captions"].apply(lambda x: bert_model.encode(x).tolist())
```
